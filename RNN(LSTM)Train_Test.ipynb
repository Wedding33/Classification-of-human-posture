```python
EPOCH = 1000
LR = 0.001   # learning rate
BATCH_SIZE = 10
```

```python
train_x=np.empty((490,128,34))
train_y=np.empty(490)
test_x=np.empty((50,128,34))#50
test_y=np.empty(50)
train_xmini=np.empty((490,128,34))
train_ymini=np.empty(10)
test_x_count=0
test_y_count=0

realtest_x=np.empty((117,128,34))
realtest_y=np.empty(117)
```

```python
for num in range(5):
    fileList = os.listdir('./data/train/00'+str(num)+'/')
    countfile=0;
    if (num<4):
        testchoose=random.sample(range(0,100),12)
    else:
        testchoose=random.sample(range(0,10),2)
        
    for filename in fileList:
        sample_path = './data/train/00'+str(num)+'/'+str(filename)
        sample = np.load(sample_path)
    
        if (sample[0,0,:,0,0].size>128):
            digout=sample[0,0,:,0,0].size-128
            digoutlist=random.sample(range(1,sample[0,0,:,0,0].size),digout)
            x=np.hstack((sample[0,0,0,:,0],sample[0,2,0,:,0]))
            x=x.reshape(1,-1)
            for i in range(1,sample[0,0,:,0,0].size):
                if (i not in digoutlist):
                    y=np.hstack((sample[0,0,i,:,0],sample[0,2,i,:,0]))
                    y=y.reshape(1,-1)
                    x=np.append(x,y,axis=0)
        elif (sample[0,0,:,0,0].size<128):
            comein=128-sample[0,0,:,0,0].size
            comeinlist=np.random.randint(1,sample[0,0,:,0,0].size,size=comein)
            x=np.hstack((sample[0,0,0,:,0],sample[0,2,0,:,0]))
            x=x.reshape(1,-1)
            for i in range(1,sample[0,0,:,0,0].size):
                y=np.hstack((sample[0,0,i,:,0],sample[0,2,i,:,0]))
                y=y.reshape(1,-1)
                x=np.append(x,y,axis=0)
                if i in comeinlist:
                    for j in range(Counter(comeinlist)[i]):
                        y=np.hstack((sample[0,0,i,:,0],sample[0,2,i,:,0]))
                        y=y.reshape(1,-1)
                        x=np.append(x,y,axis=0)
        elif (sample[0,0,:,0,0].size==128):
            x=np.hstack((sample[0,0,0,:,0],sample[0,2,0,:,0]))
            x=x.reshape(1,-1)
            for i in range(1,sample[0,0,:,0,0].size):
                y=np.hstack((sample[0,0,i,:,0],sample[0,2,i,:,0]))
                y=y.reshape(1,-1)
                x=np.append(x,y,axis=0)
                
                
        #print(num,' ',filename,' ',x.shape)
        train_x[num*100+countfile]=x
        
        if (num==0):
            temp=np.array((0))
            train_y[num*100+countfile]=temp
        elif (num==1):
            temp=np.array((1))
            train_y[num*100+countfile]=temp
        elif (num==2):
            temp=np.array((2))
            train_y[num*100+countfile]=temp
        elif (num==3):
            temp=np.array((3))
            train_y[num*100+countfile]=temp
        elif (num==4):
            temp=np.array((4))
            train_y[num*100+countfile]=temp
        
        if countfile in testchoose:
            test_x[test_x_count]=x
            test_x_count=test_x_count+1
            test_y[test_y_count]=temp
            test_y_count=test_y_count+1
            
        countfile=countfile+1
```

```python
class LSTMnet(nn.Module):
    def __init__(self, in_dim, hidden_dim, n_layer, n_class):
        super(LSTMnet, self).__init__()
        self.n_layer = n_layer
        self.hidden_dim = hidden_dim
        self.lstm = nn.LSTM(in_dim, hidden_dim, n_layer, batch_first=True)
        self.linear = nn.Linear(hidden_dim, n_class)
 
    def forward(self, x):                  
        out, _ = self.lstm(x)              
        out = out[:, -1, :]                                                  
        out = self.linear(out)             
        return out
```

```python
model = LSTMnet(34, 64, 4, 5)             
if torch.cuda.is_available():
    model = model.cuda()
    
optimizer = torch.optim.Adam(model.parameters(), lr=LR)
criterion = nn.CrossEntropyLoss()
```

```python
train_x=torch.tensor(train_x)
train_y=torch.tensor(train_y)
test_x=torch.tensor(test_x)
test_y=torch.tensor(test_y)
train_xmini=torch.tensor(train_xmini)
train_ymini=torch.tensor(train_ymini)
train_x=train_x.clone().detach()
train_y=train_y.clone().detach()
test_x=test_x.clone().detach()
test_y=test_y.clone().detach()
train_xmini=train_xmini.clone().detach()
train_ymini=train_ymini.clone().detach()
realtest_x=torch.tensor(realtest_x)
realtest_x=realtest_x.clone().detach()
realtest_y=torch.tensor(realtest_y)
realtest_y=realtest_y.clone().detach()
realtest_x=realtest_x.float().cuda()
```

```python
model.train()
```

```python
for epoch in range(EPOCH):
    iteration=0
    for i in range(49):

        train_x=train_x.float().cuda()
        train_y=train_y.long().cuda()
        c=i*10
        train_xmini=train_x[c:c+10,:,:] 
        train_ymini=train_y[c:c+10]
        output = model(train_xmini)
        loss = criterion(output, train_ymini)  
        optimizer.zero_grad()              
        loss.backward()                    
        optimizer.step()                   
        iteration=iteration+1
        
        if (epoch % 10==0) and (iteration % 10 == 0):
            test_x=test_x.float().cuda()
            test_output = model(test_x)
            
            test_output=test_output.cpu()
            predict_y = torch.max(test_output, 1)[1]#.numpy()
            predict_y=predict_y.cpu()
            predict_y=predict_y.numpy()
            test_y=test_y.cpu()
            
            accuracy = float((predict_y == test_y.numpy()).astype(int).sum()) / float(test_y.size(0))
            print('Train: epoch:{:<2d} | iteration:{:<4d} | loss:{:<6.4f} | accuracy:{:<4.2f}'.format(epoch, iteration, loss, accuracy))
    if (epoch % 50==0):
        test_output1 = model(realtest_x)
        test_output1=test_output1.cpu()
        predict_y1 = torch.max(test_output1, 1)[1]#.numpy()
        predict_y1=predict_y1.cpu()
        predict_y1=predict_y1.numpy()
        test_y1=realtest_y.cpu()

        accuracy = float((predict_y1 == test_y1.numpy()).astype(int).sum()) / float(test_y1.size(0))
        print("---------------------------------------------------------------------")
        print('Test: accuracy:{:<4.2f}'.format(accuracy))
        print("---------------------------------------------------------------------")
```
